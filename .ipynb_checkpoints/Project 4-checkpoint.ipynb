```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from pandas_profiling import ProfileReport
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

# Import data

```python
# Read data frame historical
data = pd.read_csv('./diamonds.csv')
df = data.copy()
```

# Pre-processing

```python
def preprocessing(df):
    """
    This function converts categorical variable into dummy variables and fills the column Z with the formula
    """
    for i in ['cut', 'color', 'clarity']:
        df = df.join(pd.get_dummies(df[i], prefix = i))
        df = df.drop(columns=[i])

    df['z'] = ((df['depth']*df['x'] + df['depth']*df['y'])/2)/100
    
    df['carat'] = np.log(df['carat'])
    
    try: 
        df['price'] = np.log(df['price'])
    finally:
        return df
```

# Training model

```python
# Prepare the dataframe
df = data.copy()
df = preprocessing(df)
```

```python
#Test data to be 30% of the entire data set
x = df.drop(columns=['price'])
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Prediction
predictions = model.predict(x_test)
y_test = np.exp(y_test)
predictions = np.exp(predictions)
print(model.score(x_test, y_test), np.sqrt(mean_squared_error(predictions, y_test)))
```

```python
sns.regplot(x_test['carat'], y_test, color='grey', label = 'Actual')
sns.regplot(x_test['carat'], predictions, label = 'Predicted')
plt.legend(loc='best')
plt.show()
```

# Predict Rick's Dataframe

```python
rick_df= pd.read_csv('./rick_diamonds.csv')

df = preprocessing(rick_df.copy())

predictions = model.predict(df)

rick_df['price_predicted'] = np.exp(predictions)
```

```python
# Save file
rick_df.to_csv('rick_predicted.csv')
```
